```python
import numpy as np
import matplotlib.pyplot as plt
from Vicsek_with_communication_nutrient_diffusion3_5 import *
```

```python
#Setup 1_gaussian

#Grid and system parameters
L=31.6
rho=0.5
N=int(rho*L**2)
delta_t=0.00001
Grid_size=100
xs=np.linspace(0,L,Grid_size)
ys=np.linspace(0,L,Grid_size)



#Particle propoerties
v_0=0.1

alpha=0.5

#Parameters controlling the communication chemical
addition_rate=100*delta_t
evaporation=100*delta_t
c_D=100*delta_t

#Parameters controlling the nutrients
n_D=1000*delta_t
K=0
growth_rate=0*delta_t
comsumption_rate=1000*delta_t


#Now how long it will run 
iterations=1000
sigma=1


initial_n_concentration=[[ np.exp(   -( (x-L/2)**2+(y-L/2 )**2  )/(2*sigma**2)   ) for x in xs] for y in ys]  

pos_1,orientations_1,c_concentracion_1,n_concentracion_1=Vicsek_with_communication_nutrient_diffusion3_5(v_0, N,L,c_D,evaporation,addition_rate, n_D,growth_rate,K,comsumption_rate,alpha,  Grid_size,iterations,initial_n_concentration=initial_n_concentration )
```

```python
#For saving plots
number_of_images=4
%matplotlib inline



for j in range(number_of_images):
      i=2*(j+1)*100 
      plt.quiver(pos_1[i][:,0],pos_1[i][:,1],orientations_1[i][:,0],orientations_1[i][:,1],color='black'  )
      plt.xlabel('x')
      plt.ylabel('y')
      # Display the image with a colorbar
      im = plt.imshow(c_concentracion_1[i].T, origin='lower', extent=[0, L, 0, L])
      plt.colorbar(im, label="$c$")
      plt.savefig("Plots/setup_gausien1_iteration="+str(i)+".pdf" )
      plt.clf()
```

```python
import numpy as np
import matplotlib.pyplot as plt
number_to_plot=300

for i in range(number_to_plot):
    x, y = pos_2.transpose(1,2,0)[i][0], pos_2.transpose(1,2,0)[i][1]
    
    # Compute the jump between successive points
    dx = np.diff(x)
    dy = np.diff(y)
    
    # Identify large jumps (assumes domain wraps at 0 and L)
    mask = (np.abs(dx) > L / 2) | (np.abs(dy) > L / 2)
    
    # Split trajectory at discontinuities
    x_split = np.split(x, np.where(mask)[0] + 1)
    y_split = np.split(y, np.where(mask)[0] + 1)
    
    # Plot each segment separately
    for xs, ys in zip(x_split, y_split):
        plt.plot(xs, ys)

plt.show()

```
