```python
import numpy as np
import matplotlib.pyplot as plt
from Vicsek_with_communication_nutrient_diffusion3_5 import *
from Animate import *
```

```python
#Setup 1

#Grid and system parameters
L=31.6
rho=0.5
N=int(rho*L**2)
delta_t=0.00001
Grid_size=100

#Particle propoerties
v_0=0.5
alpha=0.5

#Parameters controlling the communication chemical
addition_rate=100*delta_t
evaporation=1*delta_t
c_D=100*delta_t

#Parameters controlling the nutrients
n_D=1000*delta_t
K=0
growth_rate=0 
comsumption_rate=1000*delta_t

#Now how long it will run 
iterations=1000
initial_n_concentration=np.ones([Grid_size,Grid_size])

pos_1,orientations_1,c_concentracion_1,n_concentracion_1=Vicsek_with_communication_nutrient_diffusion3_5(v_0, N,L,c_D,evaporation,addition_rate, n_D,growth_rate,K,comsumption_rate,alpha,  Grid_size,iterations,initial_n_concentration=initial_n_concentration )
```

```python
from matplotlib.animation import FuncAnimation

def animate(pos, ori, nutrient_concentration, L, number_of_frames, interval=1):
    fig, ax = plt.subplots(figsize=(6, 6))

    # Set axis limits
    ax.set_xlim(0, L)
    ax.set_ylim(0, L)

    # Initial heatmap of the nutrient concentration
    im = ax.imshow(nutrient_concentration[0].T, extent=[0, L, 0, L],origin="lower" )

    # Initial quiver plot for particle movement
    qv = ax.quiver(pos[0][:, 0], pos[0][:, 1], ori[0][:,0], ori[0][:,1], color="black",scale=0.1)

    # Animation function
    def update(i):
        # Update nutrient concentration heatmap
        im.set_array(nutrient_concentration[i].T)

        # Update the position and direction of particles
        qv.set_offsets(pos[i])
        qv.set_UVC(ori[i][:,0], ori[i][:,1])

        return im,          qv  
       # return qv
    anim = FuncAnimation(fig, update, frames=np.arange(1, number_of_frames), interval=interval, blit=False)

    plt.colorbar(im, label="Concentration")
  #  plt.show()
    
    return anim
```

```python

%matplotlib qt
anim=animate(pos_1,orientations_1,c_concentracion_1,L,iterations)

#anim.save("Animations/setup_1_chemical_signal.gif")
```
